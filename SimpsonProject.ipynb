```python
 def normalProbabilityDensity(x):
    constant = 1.0 / np.sqrt(2*np.pi)
    return(constant * np.exp((-x**2) / 2.0) )
```

```python
def simpson(a, b, n):
    x=list()
    fx = list()
    h=(b-a)/n
    
    i=0
    while i<=n:
        x.append(a+i*h)
        fx.append(normalProbabilityDensity(x[i]))
        i +=1
    res = 0
    
    i = 0
    while i<=n:
        if i==0 or i==n:
            res += fx[i]
        elif i%2 !=0:
            res +=4 * fx[i]
        else:
            res += 2 * fx[i]
        i+=1
    return res * (h/3)  

a = int(input("Enter UPPER LIMIT value: "))
b = int(input("Enter LOWER LIMIT value: "))
n = int(input("Enter value of n: "))
print('-----------------------')

print("SIMPSON RULE")
print('value of integral is: ',simpson(a, b, n))
```
