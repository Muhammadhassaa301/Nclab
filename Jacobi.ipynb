```python
import numpy as np 
import pandas as pd
df = pd.DataFrame(columns=['x1','x2','x3'])
c = np.array([[8,-3,2,45],[4,11 ,-1,71],[6,3,12,35]])
x = np.ones(3)

x = np.zeros(len(c))
xcop = np.zeros(len(c)) 
for k in range(0,15):
    for i in range(0,len(c)):
        y=0
        for j in range(0,len(c)):
            if i != j:
                y = y + (c[i][j] * x[j]) 
        y = c[i][len(c)] - y
        xcop[i] = y / c[i][i] 
    for i in range(0,len(x)):
        x[i] = xcop[i]
    df.loc[k+1] = [x[0],x[1],x[2]]
    
round(df,4)
```
